```python
#  part 4 add in sound after correct or wrong answer
import cv2
from scipy.io import wavfile
import sounddevice as sd
import soundfile as sf

from cvzone.HandTrackingModule import HandDetector

ColorR1 = (0,255,255)
ColorR2 = (0,255,255)
ColorR3 = (0,255,255)

cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)


detector = HandDetector(detectionCon=0.8)
updown_counter = 2
soundcounter = 2
fs, data = wavfile.read("1.wav")
sd.play(data, fs)


while True:
    
    success, img = cap.read()
    img = cv2.flip (img,1)
    hands, img = detector.findHands(img)

    if hands:
        hand1 = hands[0]
        lmList = hand1["lmList"]
        if lmList:
            cursor=lmList[8]
            if 50<cursor[0]<150 and 50<cursor[1]<150:
                ColorR1 = (0,255,0)
                updown_counter = 1
            elif 200<cursor[0]<300 and 50<cursor[1]<150:
                ColorR2 = (0,255,0)
                updown_counter = 0
            elif 350<cursor[0]<450 and 50<cursor[1]<150:
                ColorR3 = (0,255,0)
                updown_counter = 0
            else:
                ColorR1 = (0,255,255)
                ColorR2 = (0,255,255)
                ColorR3 = (0,255,255)
                updown_counter = 2
        
    cv2.rectangle(img, (50,50), (150,150),ColorR1, 3)
    cv2.rectangle(img, (200,50), (300,150),ColorR2, 3)
    cv2.rectangle(img, (350,50), (450,150),ColorR3, 3)
    cv2.putText(img,'1',(10,100), cv2.FONT_HERSHEY_PLAIN,2,(250,0,0),2)
    cv2.putText(img,'2',(270,100), cv2.FONT_HERSHEY_PLAIN,2,(250,0,0),2)
    cv2.putText(img,'3',(420,100), cv2.FONT_HERSHEY_PLAIN,2,(250,0,0),2)
        
    if soundcounter != updown_counter:
        if updown_counter == 1:
            soundcounter=updown_counter
            fs, data = wavfile.read("1.wav")
            sd.play(data, fs)
        elif updown_counter == 0:
            soundcounter=updown_counter
            fs, data = wavfile.read("buzzer.wav")
            sd.play(data, fs)
        else:
            soundcounter = updown_counter
                

    #Display
    cv2.imshow("Webcam", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        
        break

cap.release()
cv2.destroyAllWindows()
```
