```python
# multi box for drag and drop without class
import cv2
from cvzone.HandTrackingModule import HandDetector
import numpy as np


cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)
detector = HandDetector(detectionCon=0.8)

#1st drag object
colorR = (255,0,255)
cx,cy,w,h = 100,100,100,100

#2nd drag object
colorR2 = (255,0,255)
dx,dy,dw,dh = 350,100,100,100

#color for answer
colorA = (0,0,0)

while (cap.isOpened()):
    success, img=cap.read()
    img=cv2.flip(img,1)
    hands, img=detector.findHands(img)
    
    if hands:
        hand1 = hands[0]
        lmList1 = hand1["lmList"]  # List of 21 Landmark points
        cursor=lmList1[8]
        l,_,_ = detector.findDistance(lmList1[8],lmList1[12],img) # find the center point of 2 fingers
        if l<30 :
            ### code for 1st object drag
            if cx-w//2<cursor[0]<cx+w//2 and cy-h//2<cursor[1]<cy+h//2:
                colorR = (0,255,0)
                cx,cy = cursor
            else:
                colorR = (255,0,255)
            
        else:
            colorR = (255,0,255)
         
           
    #display 1st drag object
    cv2.rectangle(img,(cx-w//2,cy-h//2),(cx+w//2,cy+h//2),colorR,cv2.FILLED)
        
    #attach apple image to 1st drag object
    s_img1 = cv2.imread("apple.jpg")
    x_offset=cx-w//2 +5
    y_offset=cy-h//2 +5
    img[y_offset:y_offset+s_img1.shape[0], x_offset:x_offset+s_img1.shape[1]] = s_img1    
                
    cv2.imshow("Webcam", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
