```python
# multi box for drag and drop without class
import cv2
from cvzone.HandTrackingModule import HandDetector


cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)
detector = HandDetector(detectionCon=0.8)

#drag object
colorR2 = (255,0,255)

dx,dy,dw,dh = 350,100,100,100

while (cap.isOpened()):
    success, img = cap.read()
    img=cv2.flip(img,1)
    hands, img = detector.findHands(img)
    if hands:
        hand1 = hands[0]
        lmList1 = hand1["lmList"]  # List of 21 Landmark points
        cursor=lmList1[8]
        
        l,_,_ = detector.findDistance(lmList1[8],lmList1[12],img) # find the center point of 2 fingers
        if l<30 :
            
           
            ### code for the object drag
            if dx-dw//2<cursor[0]<dx+dw//2 and dy-dh//2<cursor[1]<dy+dh//2:
                colorR2 = (0,255,0)
                dx,dy = cursor

            else:
                colorR2 = (255,0,255)
                  
        else:
            colorR2 = (255,0,255)
    else:
        colorR2 = (255,0,255)
           
           
    #display target string
    cv2.putText(img, 'Target', (300, 290), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)
    cv2.rectangle(img,(300,300),(450,450),(255,0,255),3)

    ## check answer if box drop inside the target
    if 300<dx+dw//2<450 and 300<dy+dh//2<450:
        #display the drag object
        cv2.putText(img, 'Correct', (300, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)
        cv2.rectangle(img,(dx-dw//2,dy-dh//2),(dx+dw//2,dy+dh//2),colorR2,cv2.FILLED)

    else:
        #display the drag object
        cv2.putText(img, '', (300, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)
        cv2.rectangle(img,(dx-dw//2,dy-dh//2),(dx+dw//2,dy+dh//2),colorR2,3)
    
    

    cv2.imshow("Webcam", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
